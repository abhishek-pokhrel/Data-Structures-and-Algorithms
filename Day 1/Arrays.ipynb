```python
import matplotlib.pyplot as plt
import numpy as np

# Visualization function for array operations
def visualize_array_operations():
    # Initial array (dynamic in Python)
    array = [10, 20, 30, 40]
    
    plt.figure(figsize=(8, 6))

    # Static array visualization (Conceptual)
    plt.subplot(3, 2, 1)
    static_array = np.array([10, 20, 30, 40])
    plt.bar(range(len(static_array)), static_array, color='blue')
    plt.title('Static Array (Fixed Size)')
    plt.xlabel('Index')
    plt.ylabel('Value')

    # Dynamic array visualization
    plt.subplot(3, 2, 2)
    plt.bar(range(len(array)), array, color='green')
    plt.title('Dynamic Array (Python List)')
    plt.xlabel('Index')
    plt.ylabel('Value')

    # Insertion operation visualization
    plt.subplot(3, 2, 3)
    array.insert(2, 25)  # Insert 25 at index 2
    plt.bar(range(len(array)), array, color='orange')
    plt.title('After Insertion (25 at index 2)')
    plt.xlabel('Index')
    plt.ylabel('Value')

    # Popping operation visualization
    plt.subplot(3, 2, 4)
    array.pop(3)  # Pop element at index 3
    plt.bar(range(len(array)), array, color='purple')
    plt.title('After Popping (Element at index 3)')
    plt.xlabel('Index')
    plt.ylabel('Value')

    # Deletion operation visualization
    plt.subplot(3, 2, 5)
    del array[1]  # Delete element at index 1
    plt.bar(range(len(array)), array, color='red')
    plt.title('After Deletion (Element at index 1)')
    plt.xlabel('Index')
    plt.ylabel('Value')

    plt.tight_layout()
    plt.show()

visualize_array_operations()
```

```python
## Method 1: BruteForce Method
# Time Complexity = 0(nlogn), Space Complexity = 0(n)

def sorted_squared(array):
    n = len(array)
    res = [0]*n
    for i in range(n):
        res[i] = array[i] ** 2
    res.sort()
    return res
```

```python
# Graph

import time
import numpy as np
import matplotlib.pyplot as plt

# Define the brute-force method
def sorted_squared(array):
    n = len(array)
    res = [0]*n
    for i in range(n):
        res[i] = array[i] ** 2
    res.sort()
    return res

# Measure execution time for different input sizes
input_sizes = np.arange(1, 1001, 100)  # Array sizes from 1 to 1000, incremented by 100
times = []

for size in input_sizes:
    test_array = np.random.randint(-1000, 1000, size)  # Generate a random array of 'size' elements
    start_time = time.time()  # Start timing
    sorted_squared(test_array)  # Call the function
    end_time = time.time()  # End timing
    times.append(end_time - start_time)  # Calculate the elapsed time and store it

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(input_sizes, times, marker='o', color='blue', label='Brute-force Method')
plt.title('Time Complexity of sorted_squared Brute-force Method')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
## Method 2: Double Pointer
# Time Complexity: 0(n), Space Complexity: 0(n)

def sorted_squared(array):
    n = len(array)
    i,j = 0, n-1
    res = [0]*n
    for k in reversed(range(n)):
        if array[i]**2 > array[j]**2:
            res[k] = array[i]**2
            i+=1
        else:
            res[k] = array[j]**2
            j-= 1
    return res
```

```python
# Graph

import time
import numpy as np
import matplotlib.pyplot as plt

# Double-pointer method for sorted squared array
def sorted_squared(array):
    n = len(array)
    i, j = 0, n - 1
    res = [0] * n
    for k in reversed(range(n)):
        if array[i]**2 > array[j]**2:
            res[k] = array[i]**2
            i += 1
        else:
            res[k] = array[j]**2
            j -= 1
    return res

# Measure execution time for different input sizes
input_sizes = np.arange(1, 1001, 100)  # Array sizes from 1 to 1000, incremented by 100
times = []

for size in input_sizes:
    test_array = np.random.randint(-1000, 1000, size)  # Generate a random array of 'size' elements
    start_time = time.time()  # Start timing
    sorted_squared(test_array)  # Call the double-pointer method
    end_time = time.time()  # End timing
    times.append(end_time - start_time)  # Calculate the elapsed time and store it

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(input_sizes, times, marker='o', color='green', label='Double-Pointer Method')
plt.title('Time Complexity of sorted_squared Double-Pointer Method')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def monotonic_array(array):
    n = len(array) 
    first=array[0]
    last=array[n-1]

    if first > last:
        # monotonic decreasing
        for i in range(len(array)-1):
            return array[i] > array[i+1]

    elif first==last:
        # monotonic constant
        for i in range(len(array)-1):
            return array[i] == array[i+1]

    else:
        # monotonic increasing
        for i in range(len(array)-1):
            return array[i] < array[i+1]
    return False
```

```python
#Graph


# The monotonic_array function
def monotonic_array(array):
    n = len(array) 
    first = array[0]
    last = array[-1]

    if first > last:
        # monotonic decreasing
        for i in range(len(array)-1):
            return array[i] > array[i+1]

    elif first == last:
        # monotonic constant
        for i in range(len(array)-1):
            return array[i] == array[i+1]

    else:
        # monotonic increasing
        for i in range(len(array)-1):
            return array[i] < array[i+1]
    return False

# Test arrays for visualization
arrays = {
    "Increasing": [1, 2, 3, 4, 5, 6, 7],
    "Decreasing": [7, 6, 5, 4, 3, 2, 1],
    "Constant": [4, 4, 4, 4, 4, 4, 4],
    "Non-monotonic": [1, 3, 2, 5, 4, 7, 6]
}

# Visualizing the algorithm's decision process
plt.figure(figsize=(8, 6))

for i, (label, arr) in enumerate(arrays.items(), start=1):
    plt.subplot(2, 2, i)
    plt.plot(arr, marker='o', linestyle='-', color='blue')
    plt.title(f"Array: {label}")
    plt.xlabel('Index')
    plt.ylabel('Value')

    # Check monotonicity and highlight the decision points
    if monotonic_array(arr):
        if arr[0] < arr[-1]:
            plt.fill_between(range(len(arr)), arr, color='green', alpha=0.3, label='Monotonic Increasing')
        elif arr[0] > arr[-1]:
            plt.fill_between(range(len(arr)), arr, color='red', alpha=0.3, label='Monotonic Decreasing')
        else:
            plt.fill_between(range(len(arr)), arr, color='yellow', alpha=0.3, label='Monotonic Constant')
    else:
        plt.fill_between(range(len(arr)), arr, color='gray', alpha=0.3, label='Non-Monotonic')

    plt.legend()

plt.tight_layout()
plt.show()
```
